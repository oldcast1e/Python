```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
```

```python
data1 = [
    [141, 164, 173],
    [180, 176, 178],
    [179, 179, 179],
    [178, 178, 180],
    [179, 180, 180]
]

df1 = pd.DataFrame(data1)
```

```python
for i in range(df1.shape[1]):
    plt.plot(df1.index, df1[i], label=f"Index {i}")
plt.legend()
plt.show()
```

```python
# 초기 행렬 A와 B
A = np.array([
    [141, 164, 173],
    [180, 176, 178],
    [179, 179, 179],
    [178, 178, 180],
    [179, 180, 180]
])

B = np.array([
    [142, 164, 174],
    [180, 176, 178],
    [179, 179, 179],
    [178, 178, 180],
    [178, 180, 179]
])
```

```python
# 행렬 A와 B의 각 행에 대한 변화를 추적
rows = A.shape[0]
cols = A.shape[1]

for i in range(rows):
    plt.plot(range(cols), A[i], label=f'A[{i}]')
    plt.plot(range(cols), B[i], label=f'B[{i}]')

plt.xlabel('Column Index')
plt.ylabel('Value')
plt.title('Change of Rows in Matrices A and B')
plt.legend()
plt.grid(True)
plt.show()
```
